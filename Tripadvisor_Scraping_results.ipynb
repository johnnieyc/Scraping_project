```python
import numpy as np
import pandas as pd
```

```python
newtripadvisor1=pd.read_csv('newtripadvisor1.csv', index_col=0) # use the first column as index
newtripadvisor2=pd.read_csv('newtripadvisor2.csv', index_col=0) # use the first column as index
newtripadvisor3=pd.read_csv('newtripadvisor3.csv', index_col=0) # use the first column as index
```

```python
newtripadvisor12=pd.concat([newtripadvisor1, newtripadvisor2], axis = 0)
```

```python
newtripadvisor123=pd.concat([newtripadvisor12, newtripadvisor3], axis = 0)
```

```python
print(len(newtripadvisor123))
newtripadvisor= newtripadvisor123.groupby('hotel').first()
len(newtripadvisor)
```

```python
newtripadvisor['hotel_star'] = list(map(lambda s: int(s[0][0]),newtripadvisor['hotel_star']))
newtripadvisor['hotel_star']
```

```python
newtripadvisor['number_of_review']=list(newtripadvisor['number_of_review'].apply(lambda s: int(s[:-8].replace(',',''))))
```

```python
newtripadvisor['rank']=list(newtripadvisor['rank'].apply(lambda b: b[:4].replace('#','')))
newtripadvisor['rank']=list(newtripadvisor['rank'].apply(lambda a: a.replace('o','')))
```

```python
# Delete redundant reviews
```

```python
import plotly
# connected=True means it will download the latest version of plotly javascript library.
plotly.offline.init_notebook_mode(connected=True)
```

```python
from plotly.graph_objs import Figure, Histogram, Layout

min_ = newtripadvisor['price'].min()
max_ = newtripadvisor['price'].max()

data = [Histogram(x=newtripadvisor['price'], 
                  xbins=dict(start=min_,
                             end=max_,
                             size=(max_-min_)/10))]
layout = Layout(title="Histogram of Price",
                bargap=0.2)
fig = Figure(data=data, layout=layout)

plotly.offline.iplot(fig, show_link=False, image_width=600, image_height=400)
```
